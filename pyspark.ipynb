```python
import pandas as pd

# Charger le dataset
df = pd.read_csv("D:/projet_big_data/E-commerce Customer Behavior.csv")

# Nettoyage
df['Discount Applied'] = df['Discount Applied'].astype(bool)
df['Gender'] = df['Gender'].str.capitalize()
df['Membership Type'] = df['Membership Type'].str.capitalize()
df['Satisfaction Level'] = df['Satisfaction Level'].str.capitalize()

# Remplacer les valeurs manquantes par des valeurs par défaut
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Total Spend'].fillna(0, inplace=True)
df['Items Purchased'].fillna(0, inplace=True)
df['Average Rating'].fillna(0, inplace=True)
df['Days Since Last Purchase'].fillna(df['Days Since Last Purchase'].median(), inplace=True)

# Exporter le CSV nettoyé
df.to_csv("D:/projet_big_data/cleaned_data.csv", index=False)
print("Dataset nettoyé et prêt pour ELK !")
```
